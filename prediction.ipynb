```python
#Import neccessary libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
```

```python
#Load the model 
model = load_model('simple_rnn_imdb.h5')
model.summary()
```

```python
model.get_weights()
```

```python
#store word_index and reversed_word_index as it will use to decode the review 
word_index = imdb.get_word_index()
reversed_word_index = {value : key for key,value in word_index.items()}
```

```python
#Helper function 
#Function to decode reviews 
def decode_review(encoded_review):
    return ' '.join([reversed_word_index.get(i-3,'?') for i in encoded_review])

#Function to preprocess review input text given by user 
def preprocess_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word,2) + 3 for word in words]
    #in model we are taking input of 500 words 
    padded_review = sequence.pad_sequences([encoded_review], maxlen=500)
    return padded_review
```

```python
#Predicton function 
def predict_sentiment(review):
    preprocessed_review = preprocess_text(review)
    prediction = model.predict(preprocessed_review)
    sentiment = 'Positive' if prediction[0][0] >= 0.5 else 'Negative'
    return sentiment, prediction[0][0]
```

```python
#User input and prediction 
#Example review for testing 
example_review = "The movie was fantastic! I really loved it and would watch it again."
sentiment, score = predict_sentiment(example_review)
print(f"Review: {example_review}\nPredicted Sentiment: {sentiment} (Score: {score:.4f})")
```
