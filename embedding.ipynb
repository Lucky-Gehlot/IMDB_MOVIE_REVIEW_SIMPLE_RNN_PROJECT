```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
#actually this one hot encoding works in hashing format it means , it word is given a specific index or mailbox 
```

```python
one_hot_repr = [one_hot(words,voc_size) for words in sent]
one_hot_repr
```

```python
##Word embedding 
from tensorflow.keras.layers import Embedding

#pad sequunecs used to convert evry vector to same length so that it can processed further in matrix format 
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
sent_length=8
#pre padding means in starting 0 will add 
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
#feature representation 
voc_size = 10000
dim = 10
sent_length = 8

model = Sequential()
# By adding input_length, Keras can calculate the shapes immediately
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.compile('adam','mse')
model.summary()
#model.compile will configures the training process of the neural network , this will tell two
#main things 1. how to update weights 2. How to measure error 
#adam - optimizer adam(Adaptive Moment Estimation)
```

```python
voc_size = 10000
dim = 10
sent_length = 8

model = Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.compile('adam', 'mse')

# --- THE FIX ---
# Create a "Dummy" input: 1 sequence of 8 zeros
# This forces Keras to 'Build' the internal weight matrix
dummy_data = np.zeros((1, sent_length))
model.predict(dummy_data, verbose=0)

# Now check the summary
model.summary()
```
